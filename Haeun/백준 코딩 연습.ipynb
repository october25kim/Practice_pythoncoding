```python
def alram(H,M):
    if M>=45:
        M2 = M-45
        print(H,"시",M2,"분에 알람이 설정되었습니다.")
    elif M<45:
        H2 = H - 1
        M2 = 60 - (45 - M)
        print(H2,"시",M2,"분에 알람이 설정되었습니다.")
        
alram(5,7)        
```

```python
H, M = list(map(int, input().split(",")))

alarm = H * 60 + M - 45

if(alarm < 0):
        alarm = 23 * 60 + 60 + alarm

alarm_H = alarm // 60 #몫 
alarm_M = alarm % 60  #나머지 

print('%d %d' % (alarm_H, alarm_M))
```

```python
#더 간단한 코드 
number_set = set(range(1, 10001))
generator = set()

for i in range(1, 10001):
    for j in str(i):
        i += int(j)
    generator.add(i)
    
self_numbers = sorted(number_set - generator)
for i in self_numbers:
    print(i)
```
